```python
# Importing all the necessary files

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris=load_iris()                                          #Creating the database object
df=pd.DataFrame(iris.data,columns=iris.feature_names)     #Creating the dataframe 
y=iris.target
df.head()
```

```python
y
```

```python
# Training the Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier              #Importing the necessary files
clf=DecisionTreeClassifier(max_depth=5).fit(df,y)            #Creating the classifier object
print('Classifier Created')
```

## Visualising the DECISION TREE

```python
from sklearn.tree import plot_tree                                             #importing plot_tree for plotting the 
                                                                               #decision_tree
plt.figure(figsize=(23,19))                                                    #Specifying the length and breadth
plot_tree(clf, feature_names=iris.feature_names,rounded=True, fontsize=15)     
plt.show()
```
